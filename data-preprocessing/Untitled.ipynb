### Visualization

```python
df['Maker'].value_counts().plot(kind='barh',color = 'cornflowerblue')
plt.xlabel('Vehicles divided based on Maker')
plt.show()
```

```python
sns.histplot(data=df,x='Maker',discrete=True)
plt.subplots_adjust(right=4.5)
plt.title('Vehicles divided based on Maker')
plt.show()
```

```python
sns.displot(data=df, x = 'Boot Space',kde=True)
plt.xlabel('Distribution of data based on boot space')
```

```python
sns.histplot(data=df, x = 'Fuel',hue="Fuel",discrete= True)
plt.subplots_adjust(right=2)
plt.xlabel('Vehicle divided based on Fuel type')
```
